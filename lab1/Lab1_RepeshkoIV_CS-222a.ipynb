### Variant 2. Investigation of tank losses
1. Create a function that takes a file name and a column name as a parameter and returns a numpy array with column values.
2. Using the function from the previous task, create a numpy array of the values of the ‘tank’ column.
3. Find the daily losses of tanks.
4. Find the 4 largest daily losses of tanks.
5. Determine how many tanks were destroyed in the spring of 2023.
6. Find the average number of tanks destroyed in the last 100 days of the war.
7. Create a graph of tank losses for the past year (starting from the last date in the dataset).

```python
"""
Task 1. Create a function that takes a file name and a column name as a parameter 
and returns a NumPy array with the column values.
"""
import numpy as np
import csv

FILE_PATH = 'russia_losses_equipment.csv'
COLUMN_NAME = 'tank', 'date'


def extract_csv_data(file_path, column_name):
    """
    Extracts data from a CSV file based on the specified column.

    Args:
        file_path (str): The path to the CSV file.
        column_name (str): The name of the column to extract data from.

    Returns:
        numpy.ndarray: A NumPy array containing the values from the specified column.
                       If the column is 'date', the array contains `datetime64[D]` values;
                       otherwise, it contains integers.
    """
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)
        column_index = header.index(column_name)
        data_list = [np.datetime64(row[column_index]) if column_name == 'date' else int(row[column_index]) for row in reader]

    return np.array(data_list, dtype='datetime64[D]' if column_name == 'date' else int)
```

```python
"""
Task 2. Using the function from the previous task, 
create a NumPy array from the values of the ‘tank’ column in the csv file.
NumPy array is sorted in chronological order.
"""
losses = extract_csv_data(FILE_PATH, COLUMN_NAME[0])[::-1]

losses
```

```python
"""
Task 3. Calculate the daily losses of tanks.
"""
daily_losses = np.diff(losses)

daily_losses
```

```python
"""
Task 4. Find the four largest daily tank losses.
"""
largest_daily_losses = np.sort(daily_losses)[-4:][::-1]

largest_daily_losses
```

```python
"""
Task 5. Calculate the number of tanks destroyed in the spring of 2023.
"""
dates = extract_csv_data(FILE_PATH, COLUMN_NAME[1])[::-1]
start_index, end_index = np.where(np.isin(dates, [np.datetime64('2023-06-01'), np.datetime64('2023-09-01')]))[0]

"""
Approach 1: Calculation of losses for the period 
as the difference between the cumulative losses at the beginning and end of the period.
"""
spring_2023_losses_a1 = losses[end_index] - losses[start_index]

"""
Approach 2: Calculation of losses for the period 
as the sum of daily losses during the period.
"""
spring_2023_losses_a2 = np.sum(daily_losses[start_index:end_index])

spring_2023_losses_a1, spring_2023_losses_a2
```

```python
"""
Task 6. Calculate the arithmetic average value of destroyed tanks over the last 100 days of the war.
"""
last_100_days_avg_losses = np.mean(daily_losses[-100:])

last_100_days_avg_losses
```

```python
"""
Task 7. Create a graph of tank losses over the past year 
(starting from the last date in the dataset).
"""
import matplotlib.pyplot as plt


def plot_graph(losses, dates):
    """
    Plots a graph of daily tank losses and saves it as an image file.

    Args:
        losses (numpy.ndarray): Array of daily tank losses.
        dates (numpy.ndarray): Array of dates corresponding to the losses.

    Returns:
        None: The function saves the graph as 'russia_tank_losses.png'.
    """
    text_color = 'midnightblue'
    title_font = {'family': 'serif',
                   'color': text_color,
                   'weight': 'bold',
                   'size': 20}
    text_font = {'family': 'serif',
                  'color': text_color,
                  'weight': 'normal',
                  'size': 16}
    x = dates
    y = losses

    plt.figure(figsize=(18, 7), 
               dpi=1000, 
               facecolor='aliceblue', 
               edgecolor=text_color, 
               linewidth=3.0)
    
    plt.plot(x, y, 
             color=text_color, 
             label='Daily tank losses', 
             linestyle='dashed')

    plt.title(label='Daily losses of Russian tanks over the last year (365 days) in the 2022 Russia Ukraine War',
              fontdict=title_font, 
              loc='center', 
              pad=20)
    plt.ylabel(ylabel='Tank Losses', 
               fontdict=text_font, 
               labelpad=15, 
               loc='center')
    plt.xlabel(xlabel='Date', 
               fontdict=text_font, 
               labelpad=15, 
               loc='center')
    plt.legend(loc='upper center', 
               labelcolor=text_color)

    plt.grid(color='lightsteelblue', 
             linestyle='--', 
             linewidth=0.5)

    y_min = np.min(y)
    y_max = np.max(y)
    plt.ylim(y_min - 10, 
             y_max + 10)

    x_max = x[np.where(y == y_max)]
    plt.annotate(f'max loss = {y_max}',
                 xy=(x_max, y_max),
                 xytext=(x_max + 10, y_max - 5),
                 arrowprops=dict(facecolor='aliceblue',
                                 shrink=0.05),
                 annotation_clip=False,
                 color=text_color)

    xticks = x[np.char.endswith(x.astype(str), '-01')]
    plt.xticks(ticks=xticks, 
               color=text_color)
    plt.yticks(color=text_color)

    plt.savefig('russia_tank_losses_over_year.png')


plot_graph(daily_losses[-365:], dates[-365:])
```
